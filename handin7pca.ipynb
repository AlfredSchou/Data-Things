# Simulate sensor data & apply PCA

In this exercise, you will do the following:
1. Simulate sensor data
2. Apply PCA to simulated data
3. Simulate a new point
4. Find out which region the new point belongs to

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Step 1 Simulate synthetic predictive maintenance dataset
np.random.seed(42)
time_steps = 10000

sensors = {}
for i in range(1, 21):
    sensors[f'Sensor{i}'] = np.linspace(50, 100, time_steps) + np.random.normal(0, 20, time_steps)

# Combine into a DataFrame
data = pd.DataFrame(sensors)

print("Sample Data:")
print(data.head())
```

```python
# Step 2: Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)

# Display scaled data
print("Scaled Data Sample:")
print(pd.DataFrame(scaled_data, columns=data.columns).head())
```

Do the following tasks:
1. Apply PCA with up to 5 components
2. Visualize how much variance of the dataset is defined by each component
3. Visualize the first 2 principal components and color them by timestep
4. Simulate a new position using code below
5. Plot the new point in the original 2D PCA graph

1. Apply PCA with up to 5 components

```python
pca = PCA(n_components=5)
pca_transformed = pca.fit_transform(scaled_data)

pca_df = pd.DataFrame(pca_transformed, columns=[f'PC{i+1}' for i in range(5)])
```

2. Visualize how much variance of the dataset is defined by each component

```python
# Step 4: Explained variance plot
plt.figure(figsize=(8,5))
plt.plot(range(1, 6), pca.explained_variance_ratio_, marker='o', linestyle='--')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.title('Explained Variance by Principal Components')
plt.show()
```

3. Visualize the first 2 principal components and color them by timestep

```python
# Step 5: Visualize the first 2 principal components
plt.figure(figsize=(8,6))
plt.scatter(pca_df['PC1'], pca_df['PC2'], c=range(time_steps), cmap='viridis', alpha=0.5)
plt.colorbar(label='Timestep')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('First Two Principal Components Colored by Timestep')
plt.show()
```

4. Simulate a new position using code below

```python
# Step 6: Simulate a new position
new_sensor_values = np.array([np.random.normal(75, 10, 20)])  # Simulated new point
new_sensor_values_scaled = scaler.transform(new_sensor_values)  # Standardize new point
new_pca_values = pca.transform(new_sensor_values_scaled)[:, :2]  # Transform into first 2 PCs
print(new_pca_values)
```

5. Plot the new point in the original 2D PCA graph

```python
# Step 7: Plot new point in the original 2D PCA graph
plt.figure(figsize=(8,6))
plt.scatter(pca_df['PC1'], pca_df['PC2'], c=range(time_steps), cmap='viridis', alpha=0.5, label='Original Data')
plt.scatter(new_pca_values[:, 0], new_pca_values[:, 1], color='red', s=100, label='New Point')
plt.colorbar(label='Timestep')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Projection with New Data Point')
plt.show()
```
